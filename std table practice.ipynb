```python
from pyspark.sql import functions as f
from pyspark.sql.window import Window

df = spark.table('playground_1.mydb.amazon_sales_dataset')
df = df.select('product_category','price')

part_ = Window.partitionBy('product_category').orderBy(f.col('price').desc())

df= df.withColumn('rn', f.rank().over(part_)).filter(f.col('rn')<=5)
display(df)

```

```python
from pyspark.sql import functions as f
from pyspark.sql.window import Window

df = spark.table('playground_1.mydb.amazon_sales_dataset')
df = df.select('product_category','price')

part_ = Window.partitionBy('product_category').orderBy(f.col('price').desc())

df= df.withColumn('rn', f.row_number().over(part_))
df= df.withColumn('drank', f.dense_rank().over(part_)).filter(f.col('drank')<=5)
df = df.withColumn('avg_p', f.round(f.avg('price').over(Window.partitionBy('product_category')),2))
display(df)
```
