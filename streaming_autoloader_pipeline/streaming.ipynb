```python

csv_schema = "id INT, gender STRING, age INT, city STRING, degree STRING"




df_stream = spark.readStream \
    .format("csv") \
    .option("header", True) \
    .schema(csv_schema) \
    .load("/Volumes/playground_1/mydb/volume_1/streaming_data/")



```

```python
query = df_stream.writeStream \
    .format("delta") \
    .outputMode("append") \
    .option("checkpointLocation",
            "/Volumes/playground_1/mydb/volume_1/stream_checkpoint/checkpoint1") \
    .trigger(once=True) \
    .toTable("playground_1.mydb.student_stream")
```
